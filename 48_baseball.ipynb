```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


filename = r'c:/Users/danddi/my_ws/edu/edudata/batter_stats_2017.csv'
filename2= r'c:/Users/danddi/my_ws/edu/edudata/picher_stats_2017.csv'

batter = pd.read_csv(filename)
picher = pd.read_csv(filename2)
batter.info()
```

```python
batter.set_index('선수명',inplace=True)
```

```python
batter_team_df= batter.groupby('팀명')[['연봉(2018)']].mean()
```

```python
batter_team_mean =batter_team_df['연봉(2018)']
batter_team_mean
```

```python
total_earn_mean = batter['연봉(2018)'].mean()
total_earn_mean
```

```python
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

team = batter_team_mean.index.tolist()
team.append('평균')

x_pos = np.arange(len(team))

batter_team = batter_team_mean.tolist()
batter_team.append(total_earn_mean)

bar_list = plt.bar(x_pos, batter_team, align='center', alpha=0.5)
bar_list[0].set_color('blue')
bar_list[4].set_color('blue')
bar_list[7].set_color('blue')
bar_list[9].set_color('blue')
bar_list[len(team)-1].set_color('orange')

plt.plot([0., 10], [total_earn_mean, total_earn_mean], 'k--')
plt.xticks(x_pos, team)

plt.ylabel('연봉', rotation=0)

plt.title('팀별 2018년 평균 연봉')

plt.show()
```
